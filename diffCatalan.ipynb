```python
import time
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi Catalan iteratif
def catalan_iterative(n):
    dp = [0] * (n + 1)
    dp[0] = 1

    for i in range(1, n + 1):
        for j in range(i):
            dp[i] += dp[j] * dp[i - j - 1]
    return dp[n]

# Fungsi Catalan rekursif
def catalan_recursive(n):
    result = 0
    if n == 1 or n == 0:
        return 1
    else:
        for i in range(n):
            result += catalan_recursive(i) * catalan_recursive(n - i - 1)
        return result

# Mengukur waktu eksekusi dan menyimpan hasil untuk berbagai input
inputs = list(range(1, 22,2)) 
iterative_times = []
recursive_times = []
iterative_results = []
recursive_results = []

for n in inputs:
    # Waktu dan hasil untuk iteratif
    start_time = time.time()
    iterative_result = catalan_iterative(n)
    end_time = time.time()
    iterative_time = end_time - start_time
    iterative_times.append(iterative_time)
    iterative_results.append(iterative_result)

    # Waktu dan hasil untuk rekursif
    start_time = time.time()
    recursive_result = catalan_recursive(n)
    end_time = time.time()
    recursive_time = end_time - start_time
    recursive_times.append(recursive_time)
    recursive_results.append(recursive_result)

# Print hasil Catalan numbers dengan waktu eksekusi
print("Hasil:")
for n, iter_result, iter_time, rec_result, rec_time in zip(inputs, iterative_results, iterative_times, recursive_results, recursive_times):
    print(f"Catalan({n}): Iteratif = {iter_result} (Waktu: {iter_time:.6f} s), Rekursif = {rec_result} (Waktu: {rec_time:.6f} s)")

# Membuat grafik
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 8))
plt.plot(inputs, iterative_times, label='Iteratif', marker='o', linestyle='-', linewidth=2, color='blue')
plt.plot(inputs, recursive_times, label='Rekursif', marker='o', linestyle='--', linewidth=2, color='red')

plt.title('Perbandingan Waktu Eksekusi : Iteratif vs Rekursif', fontsize=16, fontweight='bold')
plt.xlabel('Ukuran Input (n)', fontsize=14, fontweight='bold')
plt.ylabel('Waktu Eksekusi (Seconds)', fontsize=14, fontweight='bold')
plt.xticks(inputs, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12, title="Method", title_fontsize=13)
plt.tight_layout()
plt.show()
```
